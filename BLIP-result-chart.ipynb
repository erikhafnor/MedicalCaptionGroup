```python
# ðŸ“Š Plotting Evaluation Results for BLIP Model

import matplotlib.pyplot as plt
import numpy as np

# Results from evaluations
bert_f1 = 0.8383
medbert_precision = 0.7236
medbert_recall = 0.6813
medbert_f1 = 0.7012
rouge1_f1 = 0.1685
rouge2_f1 = 0.0465
rougeL_f1 = 0.1516




```

```python
# ---------- 1. Bar Chart: Comparing BERTScore F1 vs MedBERTScore F1 ----------

plt.figure(figsize=(6,4))
models = ['BERTScore F1', 'MedBERTScore F1']
values = [bert_f1, medbert_f1]

plt.bar(models, values)
plt.title('F1 Score Comparison')
plt.ylabel('F1 Score')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# ---------- 2. Grouped Bar Chart: MedBERT Precision / Recall / F1 ----------

plt.figure(figsize=(6,4))
metrics = ['Precision', 'Recall', 'F1 Score']
values = [medbert_precision, medbert_recall, medbert_f1]

plt.bar(metrics, values)
plt.title('MedBERTScore Detailed Metrics')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# ---------- 3. Radar Chart: ROUGE and BERTScore together ----------

from math import pi

labels = ['BERT F1', 'ROUGE-1 F1', 'ROUGE-2 F1', 'ROUGE-L F1']
values = [bert_f1, rouge1_f1, rouge2_f1, rougeL_f1]

# Radar chart needs a circle
values += values[:1]  # close the circle
angles = [n / float(len(labels)) * 2 * pi for n in range(len(labels))]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))

ax.plot(angles, values, linewidth=2)
ax.fill(angles, values, alpha=0.3)

ax.set_thetagrids([a * 180/pi for a in angles[:-1]], labels)
plt.title('BLIP Model Metrics Overview')
plt.show()
```
